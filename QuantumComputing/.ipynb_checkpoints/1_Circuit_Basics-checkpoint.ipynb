# 1. Circuits Basics

The provided code is a demonstration of creating, visualizing, and simulating a simple quantum circuit using Qiskit, an open-source quantum computing framework developed by IBM.

## 1.1 Libraries

1. **Import Libraries**:
   - `numpy`: a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.
   - `qiskit`: an open-source quantum computing software development framework.

```python
import numpy as np
from qiskit import QuantumCircuit
```

## 1.2 Circuit Desing

### 1.2.1 Definition


   - `QuantumCircuit(1)`: This creates a quantum circuit with one quantum bit (qubit).

```python
# Create a Quantum Circuit acting on a quantum register of one qbit
circ = QuantumCircuit(1)
```

### 1.2.2 Operations
   - `circ.h(0)`: This applies a Hadamard (H) gate on qubit 0, which puts the qubit in superposition, meaning the qubit can be in both the |0⟩ and |1⟩ state at the same time.

```python
# Add a H gate on qubit 0, putting this qubit in superposition.
circ.h(0)
```

### 1.2.3 Visualization

```python
circ.draw('mpl')
```

## 1.3. Simulating

This involves setting an initial state for the qubit and then evolving (applying) the quantum circuit on this state.

### 1.3.1 Initial State
   - `Statevector.from_int(0, 2**1)`: sets the initial state of the quantum circuit to the ground state (|0⟩).

```python
from qiskit.quantum_info import Statevector

# Set the intial state of the simulator to the ground state using from_int
state = Statevector.from_int(0, 2**1)
state.draw('latex') #draw using latex
```

### 1.3.1 Evolved State
   - `state.evolve(circ)`: evolves the state by applying the quantum circuit (here it applies the Hadamard gate to the |0⟩ state).

```python
# Evolve the state by the quantum circuit
state = state.evolve(circ)

#draw using latex
state.draw('latex')
```

#### Alternative Representations

```python
from qiskit.visualization import array_to_latex

#Alternative way of representing in latex
array_to_latex(state)
```

```python
state.draw('qsphere')
```

## 1.4 Measurement

The next part of this script goes on to introduce the concept of measurements in quantum computing, and shows how to simulate the measurement process using Qiskit. Here's a breakdown of the code:

   - The code creates a new Quantum Circuit `meas` with one qubit and one classical bit.
   - The `.barrier(range(1))` command places a barrier in the circuit, which is used as a visual aid to separate different sections of the circuit. In hardware, it also prevents certain kinds of optimization across its boundary.
   - The `.measure(range(1), range(1))` operation maps the quantum measurement to the classical bits, effectively telling the system that it should store the result of the qubit's measurement in the classical bit.
   - The `qc = meas.compose(circ, range(1), front=True)` command composes the previously defined circuit (`circ`) that applies a Hadamard gate, with the measurement circuit (`meas`). The `front=True` parameter puts the measurement operations at the beginning of the circuit.

### 1.4.1 Add Measurements Operations

```python
# Create a Quantum Circuit
meas = QuantumCircuit(1, 1)
meas.barrier(range(1))
# map the quantum measurement to the classical bits
meas.measure(range(1), range(1))

# The Qiskit circuit object supports composition.
# Here the meas has to be first and front=True (putting it before)
# as compose must put a smaller circuit into a larger one.
qc = meas.compose(circ, range(1), front=True)

#drawing the circuit
qc.draw('mpl')
```

### 1.4.2 Simulating

   - This section begins by importing necessary modules for simulation and transpilation. Transpilation is the process of rewriting a given input circuit to match the topological constraints of a specific quantum device.
   - It then declares an `AerSimulator` backend. In Qiskit, a backend represents either a simulator or a real quantum computer, and is responsible for running quantum circuits and returning results.
   - The circuit is then transpiled for the simulator backend with the `transpile` function, and the transpiled circuit is run on the simulator backend. The number of times the circuit is run is determined by the parameter `shots`. In this case, it's 1024, which means the final state will be measured 1024 times.

```python
# Adding the transpiler to reduce the circuit to QASM instructions
# supported by the backend
from qiskit import transpile

# Use AerSimulator
from qiskit_aer import AerSimulator

backend = AerSimulator()

# First we have to transpile the quantum circuit
# to the low-level QASM instructions used by the
# backend
qc_compiled = transpile(qc, backend)

# Execute the circuit on the qasm simulator.
# We've set the number of repeats of the circuit
# to be 1024, which is the default.
job_sim = backend.run(qc_compiled, shots=1024)

# Grab the results from the job.
result_sim = job_sim.result()
```

### 1.4.3 Results


   - The results of the simulation are retrieved and stored in `result_sim`.
   - The counts (the results of each shot or measurement) are extracted with `result_sim.get_counts(qc_compiled)`.
   - The results are then printed and visualized using a histogram with `plot_histogram(counts)`. This histogram will show the probability distribution of the qubit collapsing to each state (|0⟩ or |1⟩) upon measurement, as determined from the 1024 shots.

```python
from qiskit.visualization import plot_histogram

counts = result_sim.get_counts(qc_compiled)
print(counts)

plot_histogram(counts)
```
