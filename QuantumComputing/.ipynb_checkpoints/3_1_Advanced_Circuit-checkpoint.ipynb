## 3.1 Advanced Circuits

In quantum computing, advanced circuits involve the utilization of various techniques and concepts to build more sophisticated and versatile quantum circuits. This section explores three key aspects of advanced circuits: opaque gates, composite gates, and parameterized circuits.

1. Opaque Gates
    Opaque gates in quantum computing are gates whose internal implementation or functionality is not explicitly disclosed. They are treated as black boxes, where only the input-output behavior is known. Opaque gates are useful when dealing with complex operations or when higher-level abstractions are needed, allowing users to focus on the input-output behavior without concerning themselves with the underlying details.

2. Composite Gates
    Composite gates are constructed by combining multiple elementary gates or other composite gates. They enable the creation of more complex quantum operations by encapsulating sequences of gates into a single gate. Sub-circuits are often used to define composite gates, which involve building a smaller circuit and converting it into an instruction that can be inserted into a larger quantum circuit. Composite gates provide a way to abstract complex operations and improve the modularity and reusability of quantum circuits.

3. Parameterized Circuits
    Parameterized circuits contain symbolic parameters that can be assigned specific values during execution or optimization. These parameters make the circuit components tunable or variable, allowing for adaptability and flexibility in quantum computations. By adjusting the parameter values, parameterized circuits can explore a range of computations and investigate various quantum states and algorithms. They are particularly valuable for tasks such as variational quantum algorithms and quantum machine learning.

By understanding and utilizing these advanced circuit techniques, quantum computing practitioners can design circuits with enhanced capabilities and explore the full potential of quantum algorithms and computations. The combination of opaque gates, composite gates, and parameterized circuits provides a powerful toolkit for constructing and manipulating complex quantum systems.

```python
import numpy  as np
import math   as mt
import qiskit as qk

from IPython.display import display, Markdown
```

## 3.1 Opaque Gates

Opaque gates in quantum computing refer to gates whose internal implementation is not explicitly defined or disclosed. They are treated as black boxes, where the only information provided is the input and output behavior of the gate. Opaque gates are often used when the specific details of the gate's operation are not relevant or when the gate represents a complex operation that is not easily decomposed into elementary gates.

In the given example code, an opaque gate named "my_gate" is defined and applied to two sets of qubits. The gate is applied between qubits 0 and 1, as well as between qubits 1 and 2, using the `append` method in the Quantum Circuit. The specific internal implementation or functionality of the `my_gate` is not provided in the code snippet.

Opaque gates are useful when dealing with complex operations or when working with higher-level abstractions in quantum computing. They allow the user to focus on the input-output behavior of the gate without concerning themselves with the underlying details of the gate's implementation.

The visualization of the quantum circuit, generated using the `draw` method, provides a graphical representation of the circuit with the opaque gates applied to the specified qubits.

```python
from qiskit.circuit import Gate

my_gate = Gate(name='my_gate', num_qubits=2, params=[])
```

```python
qb = qk.QuantumRegister(size = 3, name = 'q')
qc = qk.QuantumCircuit(qb)

qc.append(my_gate, [qb[0], qb[1]])
qc.append(my_gate, [qb[1], qb[2]])

# Visualization
display(qc.draw('mpl')) 
```

## 3.2 Composite Gates
### 3.2.1 Sub Circuit

In quantum computing, composite gates are constructed by combining multiple elementary gates or other composite gates. They allow for the creation of more complex quantum operations by encapsulating a sequence of gates into a single gate. One way to create composite gates is by defining and using sub-circuits.

In the provided example code, a sub-circuit is built using a separate quantum register (`sub_qb`) and quantum circuit (`sub_qc`). The sub-circuit is given the name "sub_circ" for reference. 

```python
# Build a sub-circuit
sub_qb = qk.QuantumRegister(2)
sub_qc = qk.QuantumCircuit(sub_qb, name='sub_circ')

# Design the sub-circuit
sub_qc.h(sub_qb[0])
sub_qc.crz(1, sub_qb[0], sub_qb[1])
sub_qc.barrier()
sub_qc.id(sub_qb[1])
sub_qc.u(1, 2, -2, sub_qb[0])

display(sub_qc.draw('mpl')) 

# Convert to a gate and stick it into an arbitrary place in the bigger circuit
sub_inst = sub_qc.to_instruction()
```

### 3.2.2 Main Circuit

In the context of composite gates, the main circuit refers to the larger quantum circuit where the sub-circuit (discussed in section 3.2.1) is incorporated as a composite gate.

```python
# Build the main circuit
qb = qk.QuantumRegister(3, 'q')
qc = qk.QuantumCircuit(qb)

# Design the circuit
qc.h(qb[0])
qc.cx(qb[0], qb[1])
qc.cx(qb[1], qb[2])
qc.append(sub_qc, [qb[1], qb[2]])

display(qc.draw('mpl')) 
```

### 3.2.3 Complete Circuit

```python
decomposed_qc = qc.decompose() # Does not modify original circuit
display(decomposed_qc.draw('mpl'))
```

## 3.3 Parameterized Circuits

Parameterized circuits are quantum circuits that contain parameters, which are symbolic variables that can be assigned specific values during the execution or optimization of the circuit. These parameters allow for the creation of circuits with tunable or variable components, enabling flexibility and adaptability in quantum computations.

```python
from qiskit.circuit import Parameter

# Parameters
theta = Parameter(name = '')
n = 5

# Building 
qb = qk.QuantumRegister(n)    # Quantum bits
cb = qk.ClassicalRegister(1)  # Classical bits
qc = qk.QuantumCircuit(qb, cb) # Quantum circuit

# Design

qc.h(0)
for i in range(n-1):
    qc.cx(i, i+1)

qc.barrier()
qc.rz(theta, range(5))
qc.barrier()

for i in reversed(range(n-1)):
    qc.cx(i, i+1)
qc.h(0)
qc.measure(0, 0)


# Visualization
display(qc.draw('mpl'))
```

### 3.2.1 Binding Parameters to Values

```python
qc_0 = qc.bind_parameters({theta: mt.pi})

# Visualization
display(qc_0.draw('mpl'))
```

```python
qc_list = circuits = [qc.bind_parameters({theta: theta_val})
                      for theta_val in np.linspace(0, 2 * np.pi, 128)]

display(qc_list[0].draw('mpl'))
```

**Note:** Compiling over a parameterized circuit prior to binding can, in some cases, significantly reduce compilation time as compared to compiling over a set of bound circuits.
