# 4. Quantum Fourier Transform (QFT)

The Quantum Fourier Transform (QFT) is a fundamental operation in quantum computing that plays a crucial role in many quantum algorithms, including Shor's algorithm for factoring large numbers. It is used to convert a quantum state in the computational basis to its frequency representation. The QFT can be represented by the unitary operator:

$$ U_{QFT} = \frac{1}{\sqrt{N}} \sum^{N-1}_{j=0}\sum^{N-1}_{k=0} e^{2\pi i \frac{jk}{N} } |k\rangle\langle j| $$

where $N$ is the dimension of the quantum state.

In this notebook, we will implement the QFT circuit using two gates: the single-qubit Hadamard gate and the $\text{UROT}_k$ gate.

```python
import qiskit as qk
import math   as mt

from qiskit import Aer
from qiskit.quantum_info import Statevector
from qiskit.quantum_info.operators import Operator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from IPython.display import display, Markdown
```

## 4.1 2-qubit Implementation
To understand the QFT, we will start by implementing it for a 2-qubit system. The circuit for the 2-qubit QFT is as follows:

![2-qubit QFT Circuit](https://learn.qiskit.org/content/v2/ch-algorithms/images/qft.png)

```python
# Build
qb = qk.QuantumRegister(2)  # Quantum bits
qc = qk.QuantumCircuit(qb)  # Quantum Circuit

# Design

# First step
qc.h(qubit = qb[0]) # Hadamard Gate
qc.cp(theta = mt.pi/2, control_qubit = qb[1], target_qubit = qb[0]) # Urot1 Control
qc.barrier()

# Second step
qc.h(qubit = qb[1]) # Hadamard Gate
qc.barrier()

#Swap step
qc.swap(qubit1 = qb[0], qubit2 = qb[-1])

display(qc.draw('mpl'))
```

## 4.2 General Implementation
After understanding the 2-qubit case, we will move on to implementing the QFT for a general number of qubits. The general QFT circuit will follow a similar structure as the 2-qubit circuit but will have more Hadamard and controlled-$\text{UROT}_k$ gates.

```python
def quantum_fourier_transform(n_qubits):
    """
    This function creates a quantum circuit to perform the Quantum Fourier Transform (QFT).

    Args:
    n_qubits (int): The number of qubits in the circuit.

    Returns:
    qiskit.circuit.quantumcircuit.QuantumCircuit: The QFT circuit.
    """
    # Build
    qubits  = qk.QuantumRegister(n_qubits) # Quantum bits
    circuit = qk.QuantumCircuit(qubits)    # Quantum Circuit

    # Design
    for i in range(n_qubits):
        # Apply the Hadamard gate to the ith qubit
        circuit.h(qubit = qubits[i])

        # Apply a sequence of controlled phase rotation gates
        for k, j in enumerate(range(i+1, n_qubits)):
            circuit.cp(theta = mt.pi/(2**(k+1)), control_qubit = qubits[j], target_qubit = qubits[i])
            
        # Barrier for better visualization
        circuit.barrier()

    # Swap the qubits
    for i in range(n_qubits//2):
        circuit.swap(qubit1 = qubits[i], qubit2 = qubits[n_qubits-i-1])

    return circuit

n_qubits = 4
qc = quantum_fourier_transform(n_qubits)
display(qc.draw('mpl'))
```

## 4.3 First Example: bin(5)
In this section, we will apply the QFT to a specific example: converting the binary representation of the decimal number 5 to its frequency representation. We will start by initializing the state and then apply the QFT circuit.
### 4.3.1 Initial State
Before applying the QFT, we need to prepare the initial state. For the binary representation of 5, the initial state will be $|5\rangle = |101\rangle$. We can represent this state in the quantum circuit using three qubits.

```python
# Number 5 in binary system
display(bin(5))

# Implement quantum circuit with a initial state of number 5.
# Build
n_qubits = 3
qb = qk.QuantumRegister(n_qubits) # Quantum bits
initial_qc = qk.QuantumCircuit(qb)    # Quantum Circuit

# Encode the state 5.
# Design
initial_qc.x(0)
initial_qc.x(2)

display(initial_qc.draw('mpl'))
```

```python
# Set the initial state of the simulator to the ground state using from_int
state_initial = Statevector.from_int(0, 2**3)

# Generate latex_source representation for initial state
latex_initial = state_initial.draw('latex_source')

# Evolve the state by the quantum circuit
state_final = state_initial.evolve(initial_qc)

# Generate latex_source representation for final state
latex_final = state_final.draw('latex_source')

# Display the states on a single line with an arrow in between
display(Markdown(f"**Evolving States:**\n${latex_initial} \\longrightarrow {latex_final}$\n"))

display(Markdown("Final State Representation:"))
display(plot_bloch_multivector(state_final))
```

### 4.3.2 Complete Circuit: Initial State + QFT
Once we have the initial state prepared, we can apply the QFT circuit to convert it to its frequency representation. The complete circuit will consist of the initial state preparation followed by the QFT circuit.

```python
sub_qc = quantum_fourier_transform(n_qubits)
# Convert to a gate and stick it into an arbitrary place in the bigger circuit
sub_inst = sub_qc.to_instruction()

initial_qc.append(sub_qc, [qb[0],qb[1],qb[2]])
qc = initial_qc.decompose() # Does not modify original circuit

display(qc.draw('mpl'))
```

```python
# Set the initial state of the simulator to the ground state using from_int
state_initial = Statevector.from_int(0, 2**3)

# Generate latex_source representation for initial state
latex_initial = state_initial.draw('latex_source')

# Evolve the state by the quantum circuit
state_final = state_initial.evolve(qc)

# Generate latex_source representation for final state
latex_final = state_final.draw('latex_source')

# Display the states on a single line with an arrow in between
display(Markdown(f"**Evolving States:**\n${latex_initial} \\longrightarrow {latex_final}$\n"))

display(Markdown("Final State Representation:"))
display(plot_bloch_multivector(state_final))
```
